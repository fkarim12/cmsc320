Part 1

Step 1

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup

# requesting data
r = requests.get('https://cmsc320.github.io/files/top-50-solar-flares.html')

root = BeautifulSoup(r.content)

temp_table = root.find('table')

# creating data frame
table = pd.DataFrame(columns=['Rank', 'X_Class', 'Date', 'Region', 'Start Time',
                              'Max Time', 'End Time', 'Movie'], 
                    index = range(0,50))

# inserting data into table
curr_row = 0
for row in temp_table.find_all('tr'):
    curr_col = 0
    columns = row.find_all('td')
    curr_row += 1
    for column in columns:
        if (curr_row > 1):
            table.iat[curr_row - 2, curr_col] = column.get_text()
            curr_col += 1

table
```

Step 2

```python
# creating data/times
start = pd.to_datetime(table['Date'] + ' ' + table['Start Time'])
max = pd.to_datetime(table['Date'] + ' ' + table['Max Time'])
end = pd.to_datetime(table['Date'] + ' ' + table['End Time'])

table['Start Date/TIme'] = start
table['Max Date/Time'] = max
table['End Date/Time'] = end

# removing old columns
table = table.drop(['Movie', 'Date', 'Start Time', 'Max Time', 'End Time'], 1)

table
```

Step 3

```python
# gathering data
r = requests.get('https://cmsc320.github.io/files/waves_type2.html')

root = BeautifulSoup(r.content)
root.prettify()

# isolating table from 'pre' tag in html file
root = root.find('pre').get_text()
root = root.split('\n')[14:482]

# creating dataframe
nasa_table = pd.DataFrame(root)
nasa_table = nasa_table[0].apply(lambda x: pd.Series(x.split()))
nasa_table = nasa_table.drop([15, 16, 17, 18, 19, 20, 21, 22, 23], 1)
nasa_table.columns = ['Start Date', 'Start Time', 'End Date', 'End Time', 
                      'Start Freq', 'End Freq', 'flare_location', 
                      'flare_region', 'flare_class', 'cme_date', 'cme_time', 
                      'cme_angle', 'cme_width', 'cme_speed', 'plot']

nasa_table
```

Step 4

```python
# changing missing entries to NaN
nasa_table.replace(['----', '-----', '------', '--/--', '--:--', \
              '????', 'DSF', 'EP','EP?', 'FILA', 'DIM', 'altr'], 'NaN', 
              inplace = True)

# creating and storing Halo column
nasa_table['Is Halo'] = np.where(nasa_table['cme_angle'] == 'Halo',
                                 'True', 'False')
nasa_table['cme_angle'].replace('Halo', 'NA', inplace = True)

# creating and storing width lower column
nasa_table['Width Lower'] = np.where(nasa_table['cme_width'].str.contains('>'),
                                     'True', 'False')
nasa_table['cme_width'] = nasa_table['cme_width'].str.replace('>','')

# replacing instances of 24:00 because to_datetime only reads up to 23:59
nasa_table['End Time'].replace('24:00', '00:00', inplace = True)

# converting dates into date/times
for i, row in nasa_table.iterrows():
    year = row['Start Date'][:5]
    
    if row['End Date'] != 'NaN' and row['End Time'] != 'NaN':
        row['End Date'] = year + row['End Date'] + ' ' + row['End Time']
        row['End Date'] = pd.to_datetime(row['End Date'])
    
    if row['cme_date'] != 'NaN' and row['cme_time'] != 'NaN':
        row['cme_date'] = year + row['cme_date'] + ' ' + row['cme_time']
        row['cme_date'] = pd.to_datetime(row['cme_date'])
  
    row['Start Date'] = row['Start Date'] + ' ' + row['Start Time']
    row['Start Date'] = pd.to_datetime(row['Start Date'])

    #row.rename(columns = {'Start Date': 'Start Date/Time', 'End Date': 'End Date/Time', 'cme_date': 'cme_date/time'}, 1)

# removing old columns
nasa_table.drop(['Start Time', 'End Time', 'cme_time'], 1)

nasa_table
```

Part 2

Question 1

```python
top50 = nasa_table.loc[nasa_table['flare_class'].astype(str).str.contains('X')].copy()

top50['top50'] = top50['flare_class']

# removing X from the column 
top50['top50'] = top50['top50'].str.replace('X', '')
top50['top50'] = top50['top50'].astype(float)

# sort in descending order
top50 = top50.sort_values('top50', ascending = False)
top50.drop('top50', axis = 1)

top50 = top50.head(50)

top50
```

When replicating the top 50 solar flares using the NASA data, I removed the X character in the flare-class column and sorted the numbers in descending order. I got similar, but not identitical results as the SpaceWeatherLive website. Though the top solar flare is the same at X28.0 among both datasets, the last one in the SpaceWeatherLive website is X2.3, whereas in my the NASA data it is X1.8. The NASA dataset had a larger range for solar flares.
